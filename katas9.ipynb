## Python

CrÃ©er un objet dictionnaire et le sauvegarder avec le module pickle. Le charger depuis un autre notebook

```python
import pickle
my_dico = {'hello': 'world', 'missy': "cat"}

# Store data (serialize)
with open('filename.pickle', 'wb') as handle:
    pickle.dump(my_dico, handle, protocol = pickle.HIGHEST_PROTOCOL) 
    #dump prend en entre un ficher
    #dumps prend quelque chose en memoire

# Load data (deserialize)
with open('filename.pickle', 'rb') as handle:
    unserialized_data = pickle.load(handle)
    
print(my_dico == unserialized_data)
unserialized_data 
```
