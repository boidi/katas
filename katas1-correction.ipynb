Coder la fonction sigmoid et vérifier qu'elle a le comportement attendu.

```python
def sigmoid(x):
    
    return 1. / (1 + np.exp(-x))



v = np.arange(-10, 10, 0.1)
print(v)
r = sigmoid(v)

plt.scatter(v, r)
plt.show()
```

Créer une classe EasyRequests permettant de faire des requête GET et POST avec des méthodes facile à utiliser. 

```python
class EasyRequests():
    """
        A compléter
    """
    
    def __init__(self):
        """
            Appelé uniquement lorsqu'un objet est créé
        """
        print('coucou')
    
    def get(self, url):
        
        return requests.get(url).content
    
    
    def post(self, url, data, headers = None):
                
        return requests.post(url, json=data, headers=headers).content
        
        
        
        
e = EasyRequests()
e.get('http://teamliquid.net')
```
