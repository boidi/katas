Coder la fonction sigmoid et vérifier qu'elle a le comportement attendu.

```python
# fonction sigmoid sig(x)= 1/1-exp(-x)
def sig(x):
    res=1+ np.exp(-x)
    return 1/res
sig(2)
v = np.arange(-10,10,0.1)
r = sig(v)
plt.scatter(v,r) # pour afficher 
plt.show()
```
