Coder une classe Ingredient avec les attributs 
- "name", 
- "sugar". 
et la méthode : 

- get_sugar()



```python
class Ingredient(object):
    
    
    def __init__(self, name, sugar):
        
        self.name = name
        self.sugar = sugar
        
    def get_sugar(self):
        return self.sugar
```

Créer une classe "Meal" (plat) avec les attributs 
- "name" : string, 
- "recipee" : string, 
- "ingredients" : liste d'objet de type Ingredient

Implementer les méthodes :
- add_ingredient()
- get_sugar()

```python
class Meal(object):
    
    def __init__(self, name, ingredients):
        self.name = name
        self.ingredients = ingredients

    def get_sugar(self):
        
        sugar = 0 
        for ingredient in self.ingredients:
            sugar += ingredient.get_sugar()
        return sugar
```

Créer l'objet plat "tarte_au_pomme" composés des ingrédients qu'il faut et afficher la quantité de sucre total

```python
apple = Ingredient("apple", 100)
butter = Ingredient("butter", 0.1)

meal = Meal("apple pie", [apple, butter])
print(meal.get_sugar())
```
